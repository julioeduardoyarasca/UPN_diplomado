***Elaborar un programa que solicite el lado de un cuadrado y calcule su area***



1.   float es para ingresar numeros con punto decimal
2.   int es para ingresar numeros enteros (sin punto decimal)



```python
lado = int(input("Ingrese el lado del cuadrado:"))
area = lado * lado
print(area)
```

```python
lado = float(input("Ingrese el lado del cuadrado:"))
area = lado * lado
print(area)
```
