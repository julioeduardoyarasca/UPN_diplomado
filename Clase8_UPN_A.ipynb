```python
import plotly.graph_objects as go
import plotly.express as px
import numpy as np

# --- Histograma base ---
hist = go.Histogram(
    x=data["Precio"],
    nbinsx=60,
    marker=dict(
        color='rgba(0, 93, 166, 0.75)',
        line=dict(color='white', width=0.8)
    ),
    opacity=0.85,
    name="Distribución de Precios",
    hovertemplate="Precio: %{x}<br>Frecuencia: %{y}<extra></extra>"
)

# --- Curva KDE (densidad suavizada) ---
from scipy.stats import gaussian_kde
kde = gaussian_kde(data["Precio"])
xs = np.linspace(data["Precio"].min(), data["Precio"].max(), 400)
ys = kde(xs)

kde_line = go.Scatter(
    x=xs,
    y=ys * (len(data["Precio"]) * (data["Precio"].max() - data["Precio"].min()) / 60),
    mode="lines",
    line=dict(color='crimson', width=3),
    name="Densidad (KDE)"
)

# --- Línea del promedio ---
mean_price = data["Precio"].mean()
mean_line = go.Scatter(
    x=[mean_price, mean_price],
    y=[0, max(ys)*5000],
    mode="lines",
    line=dict(color="gold", width=3, dash="dash"),
    name=f"Promedio: {mean_price:.0f}",
    hovertemplate="Promedio<br>Precio: %{x}<extra></extra>"
)

# --- Diseño avanzado ---
layout = go.Layout(
    title=dict(
        text="Distribución Sofisticada del Precio",
        x=0.5,
        xanchor='center',
        font=dict(size=24)
    ),
    xaxis=dict(
        title="Precio",
        gridcolor="rgba(200,200,200,0.3)",
        zeroline=False
    ),
    yaxis=dict(
        title="Frecuencia",
        gridcolor="rgba(200,200,200,0.3)",
        zeroline=False
    ),
    template="plotly_white",
    bargap=0.05,
    hovermode="closest"
)

fig = go.Figure(data=[hist, kde_line, mean_line], layout=layout)

fig.update_layout(
    updatemenus=[
        dict(
            type="buttons",
            showactive=True,
            buttons=[
                dict(label="Normal",
                     method="relayout",
                     args=[{"xaxis.type": "linear"}]),
                dict(label="Log X",
                     method="relayout",
                     args=[{"xaxis.type": "log"}])
            ],
            x=1.12,
            y=1,
        )
    ]
)

fig.show()
```
