**Operadores Lógicos**

AND: solo es verdad cuando ambos son verdad

 OR: cuando los dos son falsos es falso

NOT: negación

```python
# Operador AND
print("True and True:", True and True)
print("True and False:", True and False)
print("False and False:", False and False)
```

```python
# Operador OR
print("True or True:", True or True)
print("True or False:", True or False)
print("False or False:", False or False)
```

```python
# Operador Negación
print("not True:",not True)
print("not False:",not False)
```

**Sentencia de control selectivo**

Simple: if
Doble: if, else
Multiple: if, elif, else

```python
#SIMPLE
x = 7
if x<10:
    print("el numero es menor a 10")
    print("estamos en el cuerpo del if")
    print("importa la sangria (identación)")
print("Programa terminado")
```

```python
#DOBLE
x = 12

if x<10:
    print("el numero es menor a 10")
    print("estamos en el cuerpo del if")
    print("importa la sangria (identación)")
else:
    print("el numero es mayor a 10")
    print("estamos en el cuerpo del else")

print("Programa terminado")
```

```python
#MULTIPLE
x = 14

if x<10:
    print("el numero es menor a 10")
    print("estamos en el cuerpo del if")
    print("importa la sangria (identación)")
elif x<15:
    print("el numero es menor a 15")
    print("estamos en el cuerpo del elif")
    print("importa la sangria (identación)")
elif x<17:
    print("el numero es menor a 17")
    print("estamos en el cuerpo del elif")
    print("importa la sangria (identación)")
else:
    print("el numero es mayor a 17")
    print("estamos en el cuerpo del else")

print("Programa terminado")
```
