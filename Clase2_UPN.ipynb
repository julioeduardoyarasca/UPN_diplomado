***Ejemplos***

1. Elabore un script que solicite la cantidad de horas y dias trabajados en una semana, posteriormente calcule el salario semanal considerando que se tiene un pago de 20 soles por hora.

```python
horas = int(input("Ingrese la cantidad de horas trabajadas:"))
dias = int(input("Ingrese la cantidad de días trabajados:"))
salario = horas * dias * 20
print(salario)
```

2. Elabore un programa que solicite la base y altura de un triangulo; posteriormente calcula el área.

```python
base = float(input("Ingrese la longitud de la base:"))
altura = float(input("Ingrese la longitud de la altura:"))
area = base * altura / 2
print(area)
```

 **Ejemplo DC1**

Calcule el monto total a pagar por un producto aplicando IGV (18%).

           Fórmula: Total = precio * cantidad * 1.18

Ejemplo

Entrada:

precio = 25

cantidad = 4

Salida:

```python
precio = int(input("Ingrese el precio:"))
cantidad = int(input("Ingrese la cantidad:"))
total = precio * cantidad * 1.18
print(total)
```
