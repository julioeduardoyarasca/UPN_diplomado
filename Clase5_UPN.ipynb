```python
def promedio(lista):
    suma = 0
    for i in lista:
        suma += i
    promedio = suma / len(lista)
    return promedio
```

```python
lista = [1,2,3,4,5,6]
promedio(lista)
```

```python
def maximo(lista):
    maximo = lista[0]
    for i in lista:
        if i>maximo:
            maximo = i
    return maximo
```

```python
lista = [2,5,3,40,5,8,10,1]
maximo(lista)
```

```python
def suma_valor(d):
    suma = 0
    for i in d:
        suma+=d[i]
    return suma
```

```python
d = {'a':10,'b':20}
suma_valor(d)
```

```python
def intercambiar(diccionario):
    d1 = {}
    for i in diccionario:
        clave = i
        valor = diccionario[i]
        d1[valor]=clave
    return d1
```

```python
d = {'a':10,'b':20}
intercambiar(d)
```

```python
#Triangulo tiene como atributos base y altura
#Triangulo es capaz de calcular su area
class triangulo:
    def __init__(self,base,altura):
        self.base = base
        self.altura = altura
    def area(self):
        return self.base*self.altura/2
```

```python
base = int(input("Ingrese su base:"))
altura = int(input("Ingrese su altura:"))
x = triangulo(base,altura)
print(x.area())
```
