# Task
Summarize the average prices of apartments in Miraflores for each year.

## Filter Data for Miraflores

### Subtask:
Filter the dataset to include only departments located in 'Miraflores'.


**Reasoning**:
Filter the 'data' DataFrame to include only rows where the 'Distrito' column is 'Miraflores' and store the result in a new DataFrame named `miraflores_data`.



```python
miraflores_data = data.loc[data['Distrito'] == 'Miraflores']
```

**Reasoning**:
Now that the data has been filtered for Miraflores, the next step is to calculate the average price per year for these apartments to fulfill the main task.



```python
average_prices_miraflores_by_year = miraflores_data.groupby('Año')['Precio'].mean()
print(average_prices_miraflores_by_year)
```

### Data Analysis Key Findings
*   The highest average apartment price in Miraflores was observed in 2017, reaching \$1,253,526.0.
*   After a peak in 2017, average apartment prices in Miraflores generally show a declining trend, from \$931,709.1 in 2020 to \$524,224.8 in 2024.
*   The lowest average apartment price was recorded in 2024 at \$524,224.8, marking a significant decrease from the 2017 peak.

### Insights or Next Steps
