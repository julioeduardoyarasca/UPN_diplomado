```python
def suma_lista(x):
	suma = 0
	for i in x:
		suma += i
	return suma
```

```python
lista = [2,4,6]
suma_lista(lista)
```

```python
lista = [10,3]
suma_lista(lista)
```

```python
def contar(x):
	cantidad = 0
	for i in x:
		if i>10:
			cantidad += 1
	return cantidad
```

```python
lista = [5,12,10]
contar(lista)
```

```python
lista = [1,3]
contar(lista)
```

```python
def buscar(lista,num):
	for i in lista:
		if i==num:
			return "Encontrado"
	return "No encontrado"
```

```python
lista = [1,2,3,4]
num = 2
buscar(lista,num)
```

```python
lista = [1,2,3,4]
num = 5
buscar(lista,num)
```

***Diccionario***
